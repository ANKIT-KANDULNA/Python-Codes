```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Sample data
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([2, 3.5, 2.8, 4.6, 5.1])

# Create a linear regression model that passes through the origin
model = LinearRegression(fit_intercept=False)

# Fit the model to the data
model.fit(X, y)

# Predict
X_test = np.array([[0], [6], [7]])  # Include 0 for plotting the origin
predictions = model.predict(X_test)

# Plotting the data and the regression line
plt.scatter(X, y, color='blue')
plt.plot(X_test, predictions, color='red')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression Through the Origin')
plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

# Sample data
X = np.array([[2.5], [3.5], [5.0], [7.5], [8.0], [10.0]])
y = np.array([0, 0, 0, 1, 1, 1])  # Binary labels, 0 or 1

# Create a logistic regression model
model = LogisticRegression()

# Fit the model to the data
model.fit(X, y)

# Predict probabilities
X_test = np.linspace(0, 12, 100).reshape(-1, 1)  # Test points for plotting
probabilities = model.predict_proba(X_test)[:, 1]  # Probabilities of class 1

# Plotting the data and the logistic regression curve
plt.scatter(X, y, color='blue')
plt.plot(X_test, probabilities, color='red')
plt.xlabel('X')
plt.ylabel('Probability of class 1')
plt.title('Logistic Regression')
plt.show()
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.tree import plot_tree

# Generate sample data
X, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0,
                           n_clusters_per_class=1, random_state=42)

# Create a decision tree classifier
model = DecisionTreeClassifier()

# Fit the model to the data
model.fit(X, y)

# Plot decision tree
plt.figure(figsize=(10,6))
plot_tree(model, filled=True, feature_names=['Feature 1', 'Feature 2'], class_names=['Class 0', 'Class 1'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.neighbors import KNeighborsClassifier

# Generate sample data
X, y = make_classification(n_samples=100, n_features=2, n_redundant=0, n_clusters_per_class=1, random_state=42)

# Create KNN classifier
knn = KNeighborsClassifier(n_neighbors=3)

# Fit the model to the data
knn.fit(X, y)

# Plot decision boundary
h = 0.02  # step size in the mesh
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

# Predict the class for each data point in the mesh grid
Z = knn.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure(figsize=(10, 6))
plt.pcolormesh(xx, yy, Z, cmap='Pastel1')

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', cmap='cool')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('KNN Classifier')
plt.show()
```
