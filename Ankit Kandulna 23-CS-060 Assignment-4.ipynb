## 1. Data of a shop.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data={"Month":['January','February','March','April','May','June','July','August','September','October','November','December'],
    "Manufacturing price": [12,34,23,17,26,30,45,16,90,74,29,9],
    "Selling price":[13,45,22,57,29,35,44,65,90,15,45,13]
}
df=pd.DataFrame(data)
print(df)
print("\n")
profit=np.array(df["Selling price"] - df["Manufacturing price"])
print("Profit ",profit)
plt.plot(df["Month"],profit,marker="o",color="g")
plt.title("Data")
plt.xlabel("Month name")
plt.ylabel("profit")
plt.xticks(rotation=45)
plt.show()
```

## 2. Toothpaste selling data for a year .

## 3. Total sale for last year.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data={
    "Month":["January","February","March","April","May","June","July","August","September","October","November","December"],
    "Air Conditioner":[12,23,6,20,27,34,7,34,23,65,65,45],
    "Washing Machine":[34,58,23,12,54,34,5,45,28,7,56,32],
    "Fridge":[34,12,5,65,76,45,76,98,8,56,45,67],
    "Television":[32,121,54,3,4,66,56,4,865,65,65,87]
}
df=pd.DataFrame(data)
print(df)
T=np.sum(df["Television"])
W=np.sum(df["Washing Machine"])
F=np.sum(df["Fridge"])
A=np.sum(df["Air Conditioner"])
labels="Washing Machine","Television","Fridge","Air Conditioner"
sizes=[T,W,F,A]
colors=["pink","lightskyblue","grey","lightgreen"]
plt.pie(sizes,labels=labels,autopct='%1.1f%%',colors=colors)
plt.show()
```
